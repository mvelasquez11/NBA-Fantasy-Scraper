```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import os
import csv

os.chdir('C:\\Users\\marck.velasquez\\PycharmProjects\\Fantasy_Basketball\\Web_Scraping')
url = 'https://www.basketball-reference.com/leagues/NBA_2020_totals.html'
page = requests.get(url)
page
page.content

soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

```python
names = ['csk', 'data-append-csv']
player_info = []
player_name = [] 
player_shortcut = []
player_link = []

for link in soup.find_all('td',attrs={"data-stat":"player"}):
    player_info.append(link.get("csk").strip() + ";" + link.get("data-append-csv").strip())
    player_shortcut.append(link.get("data-append-csv").strip("'"))
    player_name.append(link.get("csk").strip())


    
# Now that we have the links in a list we can add them to the root html we are using for the current season 
start_html = 'https://www.basketball-reference.com/players/b/' 
end_html = '/gamelog/2020/'

for item in player_shortcut:
    player_link.append(start_html + item + end_html)

os.chdir('C:\\Users\\marck.velasquez\\PycharmProjects\\Fantasy_Basketball\\Web_Scraping')
with open("player_links.csv", "w", encoding='utf-8', newline='') as f_output:
    csv_output = csv.writer(f_output, dialect='excel')
    csv_output.writerows([player_link])
    

with open("player_shortnames.csv", "w", encoding='utf-8', newline='') as f_output:
    csv_output = csv.writer(f_output, dialect='excel')
    csv_output.writerows([player_shortcut])

    
with open("player_name.csv", "w", encoding='utf-8', newline='') as f_output:
    csv_output = csv.writer(f_output, dialect='excel')
    csv_output.writerows([player_name])
```
